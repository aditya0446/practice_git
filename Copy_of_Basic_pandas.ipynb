```python
#Arithmetic operators
dict={'a':[1,2,3,4],'b':[5,6,7,8]}
data=pd.DataFrame(dict)

data['c']=data['a']+data['b']
print(data)
print()

data['c']=data['a']-data['b']
print(data)
print()

data['c']=data['a']*data['b']
print(data)
print()

data['c']=data['a']/data['b']
print(data)
```
